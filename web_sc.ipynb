```python
import requests
from bs4 import BeautifulSoup
import re
import time
```

### 放送年・クールを指定して、アニメハックで使われている各アニメに飛べるリンクを取得
##### /program/108503/ こういうやつ

```python
num = list(range(1958, 2018))
numi = []
for i in range(len(num)):
    numi.append(str(num[i]) + '-spring/')
    numi.append(str(num[i]) + '-summer/')
    numi.append(str(num[i]) + '-autumn/')
    numi.append(str(num[i]) + '-winter/')
```

```python
#参考：https://yu-nix.com/blog/2020/11/25/bs4-href/#find()%E3%81%A7href%E3%82%92%E5%BE%97%E3%82%8B
print('range:'+str(len(numi)))

href = []

for i in range(len(numi)):
    #スクレイピングの間隔をあける
    time.sleep(2)
    print('time:'+str(i)+"\033[2A",end="")
    
    #基本のURL
    url = 'https://anime.eiga.com/program/season/'
    #URL合体
    load_url = url + str(numi[i])
    
    html = requests.get(load_url)
    soup = BeautifulSoup(html.content, "html.parser")
    
    # findを使ってリンク要素を取得
    link = soup.find_all('a')
    
    #存在しない場合は飛ばす
    if link == None:
        continue
    else:
        #タグに書いてあるやつを取る
        for i in link:
            href.append(i.get('href'))
```

```python
URL_lis = []
for i in range(len(href)):
    #書類の名前をsに入れる
    s = str(href[i])
    
    #正規表現で～を含む検索
    if re.findall(r'/program/[0-9]{6}/$',s):
        #含んでいればタイトルをリストに追加
        URL_lis.append(s)
```

### アニメジャンル用スクレイピング

```python
#スクレイピングの都合で同じのが重複しちゃうので、1つ飛ばしでリストに

genre_list = []
URL_lis_genre = ['https://anime.dmkt-sp.jp/animestore/gen_sel_pc']

for k in range(len(URL_lis_genre)):
    
    #スクレイピングの間隔をあける
    time.sleep(1)
    
    #基本のURL
    url = URL_lis_genre[k]
    
    html = requests.get(url)
    soup = BeautifulSoup(html.content, "html.parser")
    
    #取得
    genre = soup.find_all(class_='btnList')
    #ないときはやめておく
    if genre == None:
        continue
    else:
        for i in genre:
            genre_list.append(i.text)
```

```python
genre_list_ = []
for i in range(len(genre_list)):

    s = str(genre_list[i][2:-3])
    m = re.search(r'\(', s)
    
    ss = s[:m.start()] + '\n'
    genre_list_.append(ss)
```

```python
genre_list_new = []
for i in range(len(genre_list_)):
    #print(genre_list_[i][:-1].split('/'))
    genre_list_new.append(genre_list_[i][:-1].split('/'))
```

```python
#スクレイピング

genre_list_c = []
URL = 'https://thesaurus.weblio.jp/content/' #基本
URL_lis_genre = sum(genre_list_new,[])[:-5]

for k in range(len(URL_lis_genre)):
    genre_list_v = []
    
    #スクレイピングの間隔をあける
    time.sleep(1)
        
    #基本のURL
    url = URL + URL_lis_genre[k]
    html = requests.get(url)
    soup = BeautifulSoup(html.content, "html.parser")
    time.sleep(1)
    
    #取得
    g = soup.find_all(class_='crosslink')

    #ないときはやめておく
    if g == None:
        continue
    else:
        for v in g:
            genre_list_v.append(v.text)
            
    genre_list_c.append(genre_list_v)
```
